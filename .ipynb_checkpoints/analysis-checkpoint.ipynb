This notebook processes UCOP-provided financial data from a subset of UC campuses.  Focus is on UC Berkeley and its College of Natural resources.

```python
import pandas as pd
import numpy as np
import os
os.system('cmd command')

import matplotlib.pyplot as plt
import seaborn as sns; #sns.set()

from matplotlib.ticker import MaxNLocator
```

Environment-specific, getting set up:

```python
os.chdir('/Users/callaway/Dropbox/service/campus committees/CNR EXCOM/UCOP data/spreadsheets/')
```

```python
UCBdata = pd.read_excel('UCB Schedule C FY14-18/UCBanalysis.xlsx')
```

```python
UCBdata.head()
```

```python
# check that totals line up:

# first sum spending by fund type
tot1 = np.sum(UCBdata.loc[:,'unrestricted_general':'restricted'], axis = 1)
# the sum spending by end point
tot2 = np.sum(UCBdata.loc[:,'salaries_wages':'transfers'], axis = 1)
# now take the difference
diff_tot=tot1-tot2
print('max difference is', np.max(diff_tot))
print('min difference is', np.min(diff_tot))
diff_tot_sort = np.sort(diff_tot, axis = 0)
diff_tot_sort[-5:]
```

Those differences are small and infrequent enough not to matter.

```python
UCBdata.loc[:,'total'] = np.sum(UCBdata.loc[:,'unrestricted_general':'restricted'], axis = 1)
UCBdata.head()
```

```python
idx = (UCBdata.loc[:,'unit'] == 'AES') | (UCBdata.loc[:,'unit'] == 'CNR')
UCBdata_CNR = UCBdata.loc[idx,:]
# scale so everything is in millions of dollars
UCBdata_CNR_InstrRes.loc[:,'unrestricted_general':'total']
UCBdata_CNR_InstrRes = UCBdata_CNR.loc[(UCBdata_CNR.loc[:,'category'] =='research') 
                                      | (UCBdata_CNR.loc[:,'category'] =='instruction'),:]
UCBdata_CNR_InstrRes.loc[:,'TotM'] = UCBdata_CNR_InstrRes.loc[:,'total']
UCBdata_CNR_InstrRes
```

```python
grouped_unit = UCBdata_CNR_InstrRes.groupby(by = ['year','category'], as_index=False).sum()
grouped_unit
```

First step, plot CNR teaching, service and research expenditures over last 5 years

```python
sns.set()
sns.set_context("talk")
ax = sns.lineplot(x = 'year',
             y = 'TotM',
             hue = 'category',
             marker = 'o',
             data = grouped_unit);
ax.set(ylabel='Total (millions)')
plt.ylim(0, 60)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.ticklabel_format(style='plain', axis='y')
```

```python
ax = sns.lineplot(x = 'year',
             y = 'TotM',
             hue = 'unit',
             marker = 'o',
             data = UCBdata_CNR_InstrRes.loc[
                 UCBdata_CNR_InstrRes.loc[:,'category']=='research',:]);
ax.set(ylabel='Total (millions)')
ax.set_title('CNR research expenditures')
plt.ylim(0, 60)
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.ticklabel_format(style='plain', axis='y')
```
